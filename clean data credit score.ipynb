```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

credit_df=pd.read_csv('data/clean data.csv',low_memory=False)
credit_df
```

```python
cols_to_fix = ['Age','Annual_Income','Num_of_Loan','Num_of_Delayed_Payment',
               'Changed_Credit_Limit','Outstanding_Debt',
               'Amount_invested_monthly','Monthly_Balance']
#Some numeric features like age were object datatype
for col in cols_to_fix:
   credit_df[col] = pd.to_numeric(credit_df[col], errors='coerce')
```

```python
credit_df.describe()
```

```python
fig,ax=plt.subplots(figsize=(14,10)) #heatmap to see correlation among all features
sns.heatmap(data=credit_df.select_dtypes(include=np.number).corr(),annot=True,cmap='RdYlGn')
```

```python
credit_df['Credit_Score_Code']=credit_df['Credit_Score'].map({'Poor':0,'Standard':1,'Good':2})
```

```python
input_cols=['Age','Occupation',
       'Annual_Income', 'Monthly_Inhand_Salary', 'Num_Bank_Accounts',
       'Num_Credit_Card', 'Interest_Rate', 'Num_of_Loan',
       'Delay_from_due_date', 'Num_of_Delayed_Payment', 'Changed_Credit_Limit',
       'Num_Credit_Inquiries', 'Credit_Mix', 'Outstanding_Debt',
       'Credit_Utilization_Ratio', 'Credit_History_Age',
       'Payment_of_Min_Amount', 'Total_EMI_per_month',
       'Amount_invested_monthly', 'Payment_Behaviour', 'Monthly_Balance']
target_col='Credit_Score_Code'
```

```python
from sklearn.model_selection import train_test_split
train_set,val_set=train_test_split(credit_df,test_size=0.2,random_state=42,stratify=credit_df[target_col])
```

```python
train_inputs=train_set[input_cols].copy()
val_inputs=val_set[input_cols].copy()
train_target=train_set[target_col].copy()
val_target=val_set[target_col].copy()
```

```python
numeric_cols=train_inputs.select_dtypes(include=np.number).columns.tolist()
cat_cols=train_inputs.select_dtypes(include='object').columns.tolist()
```

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(strategy='median')
imputer.fit(train_inputs[numeric_cols])
```

```python
train_inputs[numeric_cols]=imputer.transform(train_inputs[numeric_cols])
val_inputs[numeric_cols]=imputer.transform(val_inputs[numeric_cols])
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(train_inputs[numeric_cols])
```

```python
train_inputs[numeric_cols]=scaler.transform(train_inputs[numeric_cols])
val_inputs[numeric_cols]=scaler.transform(val_inputs[numeric_cols])
```

```python
from sklearn.preprocessing import OneHotEncoder
enc=OneHotEncoder(sparse_output=False,handle_unknown='ignore')
enc.fit(train_inputs[cat_cols])
```

```python
encoded_cols=enc.get_feature_names_out(cat_cols).tolist()
train_inputs[encoded_cols]=enc.transform(train_inputs[cat_cols])
val_inputs[encoded_cols]=enc.transform(val_inputs[cat_cols])
```

```python
x_train=train_inputs[numeric_cols+encoded_cols]
x_val=val_inputs[numeric_cols+encoded_cols]
```

```python
credit_df[target_col].value_counts()
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score #Precision Recall and f1 score give a better insight on metrics due to the case of class imbalance

def model_evaluate(model):
    train_preds=model.predict(x_train)
    val_preds=model.predict(x_val)
    train_acc=accuracy_score(train_target,train_preds)
    val_acc=accuracy_score(val_target,val_preds)
    train_prec=precision_score(train_target,train_preds,average='weighted')   
    val_prec=precision_score(val_target,val_preds,average='weighted')
    train_rec=recall_score(train_target,train_preds,average='weighted')   
    val_rec=recall_score(val_target,val_preds,average='weighted')   
    train_f1=f1_score(train_target,train_preds,average='weighted')
    val_f1=f1_score(val_target,val_preds,average='weighted')
    print(f'Train Results: Acc ==> {train_acc}, Precision ==>{train_prec},Recall ==> {train_rec},F1 ==> {train_f1}')
    print(f'Val Results: Acc ==> {val_acc}, Precsion ==>{val_prec}, Recall ==> {val_rec},F1 ==>{val_f1}')
```

```python
from sklearn.linear_model import LogisticRegression #baseline model
model_log=LogisticRegression(max_iter=2000,class_weight='balanced') #class_weight='balanced' is for handling class imbalance
model_log.fit(x_train,train_target)
```

```python
model_evaluate(model_log)
```

```python
from sklearn.tree import DecisionTreeClassifier #Tree based structured choose to see performance gain
model_tree=DecisionTreeClassifier(max_depth=10,random_state=42,class_weight='balanced')
model_tree.fit(x_train,train_target)
```

```python
model_evaluate(model_tree)
```

```python
from sklearn.ensemble import RandomForestClassifier #Random Forest performs well in such type of features
model_forest=RandomForestClassifier(n_jobs=-1,n_estimators=150,max_depth=35,random_state=42,class_weight='balanced',min_samples_leaf=12)
model_forest.fit(x_train,train_target)
```

```python
model_evaluate(model_forest) #Random Forest show performance Gain when the clean data is used hence showing importance of data quality
```

```python
df=pd.DataFrame({'cols':x_train.columns,'imp':model_forest.feature_importances_}).sort_values(by='imp',ascending=False).head(10) # Importane of Features Showing contribution to Random Forest
```

```python
from xgboost import XGBClassifier
from sklearn.utils import compute_sample_weight
sample_weight=compute_sample_weight(class_weight='balanced',y=train_target)
model_xgb=XGBClassifier(n_jobs=-1,n_estimators=100,max_depth=12,random_state=42,sample_weight=sample_weight,learning_rate=0.01) # Learning Rate is choosen to be balanced to overcome severe overfitting, also same goes for max depth
model_xgb.fit(x_train,train_target)
```

```python
model_evaluate(model_xgb)
```
